##1-Use Pandas to read the Arabic_Sentiment.csv file into a DataFrame.**bold text**

```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Phase2/Week3/Task Files-20250525T141653Z-1-001/Task Files/Arabic_Sentiment.csv")
data
```

##2-Add new columns to the DataFrame for the following calculations:




###Count the number of words in each row.


```python
data['word_count'] = data['text'].str.split().str.len()
data
```

###Count the number of characters in each row.


```python
data['characters_count'] = data['text'].str.len()
data
```

###Count the number of hashtags (#) in each row.


```python
data['#_count'] = data['text'].str.count('#')
data
```

###Calculate the average word length for each row.

```python
data['average_word_length'] =(data['characters_count'] / data['word_count']).round(1)
data
```

##3-Use the replace() method along with regular expressions to clean the text in my DataFrame.

```python
data['text'] = data['text'].str.replace('_', ' ')
data['text'] = data['text'].astype(str).str.replace(r'[^\u0600-\u06FF\s]', '', regex=True)
```

##4-After cleaning the text, what is the best way to remove rows from the DataFrame that contain empty text.

```python
data['text'].replace('', pd.NA, inplace=True)
data.dropna(subset=['text'], inplace=True)
```

##5-Use NLTK to extract trigrams (N=3) from the text data in my DataFrame, and print them in descending order.

```python
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
from collections import Counter
# Download the required resource if it's missing
try:
    nltk.data.find('tokenizers/punkt_tab/english/')
except LookupError:
    print("Downloading punkt_tab resource...")
    nltk.download('punkt_tab')
    print("punkt_tab resource downloaded.")

sentences = data['text']
# Join all sentences into a single string
sentences = ' '.join(sentences)
# Tokenize the combined string into individual words
token = word_tokenize(sentences)
# Generate n-grams from the tokenized words N=3
ngrams = ngrams(token, 3)
# Count the frequency of each n-gram using Counter and sort from most freq to smallest
COUNTS = Counter(ngrams)
most_common = COUNTS.most_common()
# Display in  descending
most_common
```

##6-Create a new column in my DataFrame that contains the text without stop words.

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# Create a list of arabic stopwords
stop_words_list_arab = list(stopwords.words('arabic'))

data['text_without_stop_words'] = data['text'].apply(lambda x: " ".join(w for w in x.split() if w not in stop_words_list_arab))
data
```

##7-Create a word cloud using the text data that has been cleaned of stop words

###7.1 Setup DICT

```python
sentences = data['text_without_stop_words']
# Join all sentences into a single string
sentences = ' '.join(sentences)
# Tokenize the combined string into individual words
token = word_tokenize(sentences)
# Count the frequency of each n-gram using Counter and sort from most freq to smallest
COUNTS = Counter(token)
most_common = COUNTS.most_common(30) #descending

# Convert the most common tokens to a dictionary
most_common_dict = {k: v for k, v in most_common}

# Display the dictionary of most common tokens
print(most_common_dict)
```

###7.2 Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# Generate the word cloud
wc = WordCloud(
    background_color="white",
    font_path='/content/drive/MyDrive/Phase2/Week3/Task Files-20250525T141653Z-1-001/Task Files/Arial.ttf'
)
wc = wc.generate_from_frequencies(most_common_dict)

# Display the word cloud
plt.rcParams['figure.figsize'] = (12, 8)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")  # Hide axes
plt.show()
```
